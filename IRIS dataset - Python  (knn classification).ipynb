<strong>Importing IRIS dataset from datasets module of scikit-learn</strong>

```python
from sklearn.datasets import load_iris
```

```python
data_iris = load_iris()
```

<strong>The dataset returned is a Bunch object which is similar to Python dictionary (key-value pairs)</strong>

```python
print('Keys for IRIS dataset: \n{}'.format(data_iris.keys()))
```

The values of the keys are explained below: <br>
    1) <b>data             : NumPy array which contains data points for the features of the dataset </b><i>(In this case, sepal length,
    sepal width, petal length and petal width)</i><br>
    2) <b>target           : NumPy array which contains output value for the data points </b> <br>
    3) <b>frame            : Shape for the dataframe </b><br>
    4) <b>target_names     : Names of the target class </b> <i>(In this case it is an array of strings)</i><br>
    5) <b>DESCR            : Gives a simple description of the dataset </b><br>
    6) <b>feature_names    : Names of the feature set for the dataset </b><br>
    7) <b>filename         : Location of the data </b><br>

```python
# First 10 rows of the IRIS dataset

print('Data points of the IRIS dataset: \n{}'.format(data_iris['data'][:10]))
```

```python
print('Target value of the IRIS dataset: \n{}'.format(data_iris['target']))

# Here 0,1 and 2 denotes the values of target names respectively
```

```python
print('Frame for the IRIS dataset: \n{}'.format(data_iris['frame']))

# While loading load_iris() in second line, if parameter as_frame=True then it would return SHAPE of the dataframe
```

```python
print('Target names for the IRIS dataset: \n{}'.format(data_iris['target_names']))
```

```python
print('Description of the IRIS dataset: \n{}'.format(data_iris['DESCR']))
```

```python
print('Feature names for the dataset: \n{}'.format(data_iris['feature_names']))
```

```python
print('File path: \n{}'.format(data_iris['filename']))
```

<strong>Dividing the data into training and testing sets</strong>

```python
print('Shape of the dataset: \n{}'.format(data_iris['data'].shape))
```

```python
# Using 'train_test_split' function of scikit-learn, data can be shuffled and then divided as 
# 75% -> Training set and 25% -> Testing set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_iris['data'], data_iris['target'], random_state = 0)
```

```python
# X_train contains 75% of the dataset and X_test contains 25% of the dataset

print('X_train: {}'.format(X_train.shape))
print('X_test: {}'.format(X_test.shape))

print('y_train: {}'.format(y_train.shape))
print('y_test: {}'.format(y_test.shape))
```

<strong>Visualizing the data using pair plot</strong>

```python
import pandas as pd

data_iris_dataframe = pd.DataFrame(X_train, columns = data_iris.feature_names)
data_iris_visualize = pd.plotting.scatter_matrix(data_iris_dataframe, c=y_train, figsize=(15, 15), marker='*', hist_kwds={'bins':20},
                                       s=60, alpha=.8)
```

<strong>Applying k-nearest neighbors model to the dataset</strong>

```python
# Machine learning models for scikit-learn are implemented in Estimator classes
from sklearn.neighbors import KNeighborsClassifier

# k in the kNN refers to the number of neighbors used to calculate the output for the dataset
knn = KNeighborsClassifier(n_neighbors=1)
```

<strong>Build a model by fitting the data</strong>

```python
knn.fit(X_train, y_train)
```

<strong>Prediction the accuracy of the model using testing set</strong>

```python
prediction_output = knn.predict(X_test)
print('Accuracy :{}'.format(knn.score(X_test, y_test)))
```

<strong>It shows accuracy of 0.97. It means 97% of the time the prediction is correct.</strong>

```python
# Testing new sample

import numpy as np 
new_data = np.array([[9.7, 3, 5.0, 4]])
new_data_class = knn.predict(new_data)
print('Class for the new data: {}'.format(data_iris['target_names'][new_data_class]))
```
